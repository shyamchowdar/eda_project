```python
playername=[]
span=[]
matches_played=[]
innings=[]
not_outs=[]
runs=[]
highest_score=[]
average=[]
hundreds=[]
fiftes=[]
ducks=[]
for i in urls:
    page=requests.get(i)
    soup=BeautifulSoup(page.text)

    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[1]
        playername.append(s2)


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[2]
        span.append(s2)


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[3]
        matches_played.append(s2)


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[4]
        innings.append(s2)


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[5]
        not_outs.append(s2)


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[6]
        runs.append(s2)  


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[7]
        highest_score.append(s2)


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[8]
        average.append(s2)


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[9]
        hundreds.append(s2)


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[10]
        fiftes.append(s2)


    s1=soup.find_all('tr', class_="data1")
    for i in s1:
        s2=i.text.split('\n')[11]
        ducks.append(s2)
```
